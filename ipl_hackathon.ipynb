```python
df_sbc = df_sbc.drop('commentary', axis=1)
df_sbc
```

```python
df_sbc = df_sbc.drop(['runningScore','runningOver'], axis=1)
df_sbc
```

```python
df_sbc = df_sbc.drop(['minutes','link','shortText','captain'], axis=1)
df_sbc
```

```python
#per match score year wise
per_match_score = df_sbc.groupby(['season','match_id']).agg(Score=('runs', 'sum')).reset_index()
per_match_score
```

```python
# year wise average score
per_match_score.columns
per_match_score_season_wise = per_match_score.groupby(['season']).agg(Match_avg_score=('Score', 'mean')).reset_index()
per_match_score_season_wise
```

```python
per_match_score_season_wise.Match_avg_score = per_match_score_season_wise.Match_avg_score.astype(int)
per_match_score_season_wise.season = per_match_score_season_wise.season.astype(int)
per_match_score_season_wise
```

```python
per_match_score_season_wise.reset_index(drop=True, inplace=True)
per_match_score_season_wise
```

```python
per_match_score_season_wise = per_match_score_season_wise.set_index(pd.Index(range(2008,2024)))
per_match_score_season_wise.drop('season',axis=1,inplace=True)
per_match_score_season_wise
```

```python
per_match_score_season_wise.plot(kind='bar',xlabel="year",ylabel="Match-avg-score(inn1 + inn2)")
```

```python

batsman_data_year_wise = df_sbc.groupby(['name']).agg(({'runs': 'sum','ballsFaced':'sum'}))
batsman_data_year_wise['strike_rate']=batsman_data_year_wise['runs']*100/batsman_data_year_wise['ballsFaced']
batsman_data_year_wise.reset_index(level=0, inplace=True)
batsman_data_year_wise.reset_index(level=0, inplace=True) 
batsman_data_year_wise= batsman_data_year_wise.astype(({'runs': int}))

batsman_data_year_wise= batsman_data_year_wise.astype(({'ballsFaced': int}))

batsman_data_year_wise

batsman_strikerate=batsman_data_year_wise.loc[(batsman_data_year_wise.strike_rate>150) &(batsman_data_year_wise.runs>2000)]
batsman_strikerate.sort_values(by=['strike_rate'],ascending=False)
batsman_strikerate
```

```python
batsman_data_year_wise.to_csv('batsman_data_year_wise.csv', index=False)
```

```python
overall_batman_record=df_sbc.groupby('name').agg(({'runs':'sum','fours':'sum','sixes':'sum'})).reset_index()
top_hitter=df_sbc.groupby('name').agg(({'runs':'sum','ballsFaced':'sum','fours':'sum','sixes':'sum'})).reset_index()
top_10_batsman=overall_batman_record.sort_values(by=['runs'],ascending=False).head(10)

# overall_batman_record.season = per_match_score_season_wise.season.astype(int)
top_10_batsman.runs=top_10_batsman.runs.astype(int)
top_10_batsman.fours=top_10_batsman.fours.astype(int)
top_10_batsman.sixes=top_10_batsman.sixes.astype(int)

top_hitter['boundary_count']=top_hitter['fours']+top_hitter['sixes']
boundry_hitter=top_hitter.loc[(top_hitter.boundary_count>300)]
top_hitter['strike_rate']=top_hitter['runs']*100/top_hitter['ballsFaced']


top_hitter['hitter_value']=(top_hitter['strike_rate']*0.55/100)+(top_hitter['boundary_count']*0.45/100)
top_hitter=top_hitter.sort_values(by=['hitter_value'],ascending=False).head(10)

top_hitter.to_csv('top_hitter.csv', index=False)
top_hitter
boundry_hitter.sort_values(by=['boundary_count'],ascending=False)



# overall_batman_record.sort_values(by=[IPL_RUN])
```
