```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("points_table.csv")
```

```python
df
```

```python
team_win_loss=df.groupby('name').agg({'matchesplayed':'sum','matcheswon':'sum','matcheslost':'sum'})
team_win_loss['win_ratio']=team_win_loss['matcheswon']/team_win_loss['matchesplayed']
team_win_loss['loss_ratio']=team_win_loss['matcheslost']/team_win_loss['matchesplayed']

team_win_ratio_sort=team_win_loss.sort_values(by=['win_ratio'],ascending=False)
team_loss_ratio_sort=team_win_loss.sort_values(by=['win_ratio'])
team_win_ratio_sort.to_csv('team_win_ratio_sort.csv', index=False)
team_loss_ratio_sort.to_csv('team_loss_ratio_sort.csv', index=False)
```

```python
team_win_ratio_sort
```

```python
team_win_ratio_sort.reset_index(level=0, inplace=True) 
```

```python
team_win_ratio_sort.to_csv('team_win_ratio_sorted.csv', index=False)
team_win_ratio_sort
```

```python
team_loss_ratio_sort.reset_index(level=0, inplace=True) 
team_loss_ratio_sort.to_csv('team_loss_ratio_sorted.csv', index=False)
team_loss_ratio_sort
```

```python
venue_data=df1.groupby('venue').agg({'overs':'sum','wickets':'sum'})
venue_data= venue_data.astype({'overs': int,'wickets':int})
venue_data['over_to_wicket_ratio']=venue_data['overs']/venue_data['wickets']
venue_data_wicket_sorted=venue_data.sort_values(by=['wickets'],ascending=False)
venue_data_ratio_sorted=venue_data.sort_values(by=['over_to_wicket_ratio'],ascending=False)
#considering atleast 100 wickets
venue_data_ratio_sorted=venue_data_ratio_sorted[venue_data_ratio_sorted['wickets']>200]
venue_data_ratio_sorted.reset_index(level=0, inplace=True) 

venue_data_ratio_sorted.to_csv('venue_data_ratio_sort.csv', index=False)
```
